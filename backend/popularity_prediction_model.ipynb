```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\25471\Desktop\Python 3.12\song-recommender\song-recommender\database\dataset.csv")
df.head()
```

```python
#Encoding data
count_map = df['track_genre'].value_counts().to_dict()
df['genre_encoded'] = df['track_genre'].map(count_map)
```

```python
numeric_df = df.select_dtypes(include='number')
numeric_df.head()
```

```python
corr = numeric_df.corr().abs().drop("popularity") #finding correlation
sort = corr["popularity"].sort_values(ascending=False)
```

```python
features = numeric_df[corr.columns.to_list()].drop("popularity", axis = 1)
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = features
y = numeric_df["popularity"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
pop_model = LinearRegression()
```

```python
pop_model.fit(xtrain, ytrain)
```

```python
ypred = pop_model.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(ytest, ypred)*100
score
```
